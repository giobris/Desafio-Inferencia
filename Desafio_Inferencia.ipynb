# Fazendo o cruzamento entre as bases PHQ e HEI

```python
df = df_phq.merge(df_hei, how = 'inner', on = "SEQN")
```

```python
df.info()
```

```python
# Criando a VariÃ¡vel de escore Patient Health Questionnaire-9
df["phq9"] = df.iloc[:, 1:10].sum(axis=1)
df.head()
```

```python
# Criando a VariÃ¡vel de grupo de sintomas de depressÃ£o
# 0 (âsem sintomasâ) se phq9 < 5
# 1 (âsintomas levesâ) se 5 <= phq9 < 10
# 2 (âsintomas moderadosâ) se 10 <= phq9 < 15
# 3 (âsintomas moderadamente severosâ) se 15 <= phq9 < 19 
# 4 (âsintomas severosâ) se phq9 >= 20
df['phq_grp'] = df['phq9'].apply(
    lambda x: 
     0 if x < 5 else 
    (1 if x >= 5 and x < 10 else 
    (2 if x >= 10 and x < 15 else 
    (3 if x >= 15 and x < 19 else 
     4)))
)
df.head()
```
